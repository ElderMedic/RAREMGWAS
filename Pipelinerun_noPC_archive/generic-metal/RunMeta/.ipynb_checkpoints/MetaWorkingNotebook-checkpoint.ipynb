```python
# slightly process metal output to draw manhattan plot
import pandas as pd
import os
import numpy as np


# In[3]:

# summary_origin = pd.read_csv("/exports/reum/CKe/RunMeta/METALresult_filtered.TBL",sep='\s+')

summary_origin = pd.read_csv("/exports/reum/CKe/generic-metal/RunMeta/results/METALresult_withTACERA_refEIRA1_noMiss.TBL",sep='\s+')
# gwas_samantha = pd.read_csv("/exports/reum/CKe/generic-metal/RunMeta/gwas_version8_merged.assoc.logistic",sep='\s+')
```

```python
BETA = np.log(gwas_samantha['OR'])
gwas_samantha.insert(7,'BETA',BETA)
```

```python
OR = np.exp(summary_origin['Effect'])
summary_origin.insert(6,'OR',OR)
```

```python
SNP = gwas_samantha['SNP'].str.split(':',expand=True)[0]+":"+gwas_samantha['SNP'].str.split(':',expand=True)[1]
gwas_samantha['SNP'] = SNP
```

```python
Merged = pd.merge(summary_origin,gwas_samantha,left_on=['MarkerName'],right_on=['SNP'],how='inner',suffixes=("_latest_meta","_Samantha"))
Merged
```

```python
from scipy.stats import pearsonr
corr, pval = pearsonr(Merged['OR_latest_meta'], Merged['OR_Samantha'])
print("Pearsons correlation: %.3f, Pvalue: %.4f" % (corr, pval) )
```

```python
# plot for TBL file generated on Oct.12 in CKe/RunMeta/METALresult_filtered 2021
# Calculate the point density
# for effect sizes
x = Merged['OR_latest_meta'][:10000]
y = Merged['OR_Samantha'][:10000]

xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
 
 # Sort the points by density, so that the densest points are plotted lastidx = z.argsort()
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]

fig, ax = plt.subplots()
plt.xlim([0.8,1.2])
# plt.xlim([0,4])
plt.ylim([0.8,1.2])
plt.scatter(x, y,c=z,  s=2,cmap='Spectral')

plt.colorbar()
plt.show()
```
