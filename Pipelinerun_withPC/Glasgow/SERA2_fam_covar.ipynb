```python
def processdcID(row):
    row = str(row).replace(" ","").lstrip("0")
    return row
def processFAM(row):
    row = str(row).split("_")[1]
    return row
```

```python
def remissionDAS28(score):
    if score < 2.6:
        state = 2 # case: achieve remission
    elif score >= 2.6:
        state = 1 # control: non-remission
    else:
        state = -9
    return state
```

```python
sera2_fam[4] = merged_sera2['Gender']
sera2_fam[4] = sera2_fam[4].map({"Male":1,"Female":2}).fillna(0)
sera2_fam[5] = merged_sera2['DAS28CRP_Month_6'].map(remissionDAS28)
sera2_fam[5] = sera2_fam[5].fillna(-9)
```
